```python
import pathlib

import pandas as pd
```

```python
input_path = pathlib.Path(
    "../../data/5.converted_data/non_aggregated_data.parquet"
).resolve(strict=True)
```

```python
# read in the data
df = pd.read_parquet(input_path)
df.head()
```

```python
# drop the columns that are not needed
uneeded_columns = ["Metadata_replicate", "Metadata_side"]
df = df.drop(columns=uneeded_columns)
```

```python
# get the variance of each feature for each genotype
variances = df.groupby("Metadata_genotype").var()
variances
```

```python
# save the per feature per genotype variance data
output_path = pathlib.Path(
    "../results/non_aggregated_results/non_aggregated_per_feature_per_genotype_variance.csv"
).resolve()
variances.to_csv(output_path)
```
